**LOADING DATASET**

```python
from pyspark.sql.functions import *
from pyspark.sql import SparkSession

spark=SparkSession.builder.appName("CustomerAnalytics").getOrCreate()
df=spark.read.csv("/Volumes/likitha_catalog/default/project-1/E-commerce Dataset.csv",header=True,inferSchema=True)
df.display()
```

**DATA CLEANSING**



```python
# Remove duplicates
df = df.dropDuplicates()

# Remove nulls
df = df.na.drop()

# # Convert date column
# df = df.withColumn(
#     "Order_Date",
#     expr("try_cast(Order_Date AS DATE)")
# )

df.display()
```

**DATA TRANSFORMATION**



```python
from pyspark.sql.functions import *
#final sales after discount
df = df.withColumn("final_sales",col("sales")*(1-col("discount")))
#proft after shipping
df = df.withColumn("net_profit",col("profit")-col("shipping_cost"))
#Delivery speed Category
df = df.withColumn("Delivery_Type",when(col("aging")<=2,"Fast").when(col("aging")<=5,"Medium").otherwise("Slow"))
df.display()
```

da

```python
from pyspark.sql
total_revenue=df.agg(sum("final_sales").alias("Total_Revenue"))
total_revenue.show()
```

```python
# Sales by Product Category
from pyspark.sql.functions import *
sales_by_category=df.groupBy("Product_Category").agg(sum("final_sales").alias("total_sales"))
sales_by_category.show()
```

```python
#Top Customers
from pyspark.sql.functions import *
top_customers=df.groupBy("Customer_Id").agg(sum(col("final_sales").alias("total_sales")).orderBy(col("total_sales").desc())
top_customers.display()
```

```python
# Sales by Device Type
from pyspark.sql.functions import *
sales_by_device=df.groupBy("Device_Type").agg(sum(col("final_sales")).alias("device_sales"))
display(sales_by_device)
```

```python
# Delivery Analysis
delivery_analysis = df.groupBy("Delivery_Type") \
    .count()
delivery_analysis.show()
```

```python
#PROFIT BY CATEGORY
from pyspark.sql.functions import *
profit_by_category=df.groupBy("Product_Category").agg(sum("net_profit").alias("total_profit"))
profit_by_category.show()
```

```python
#SAVING THE OUTPUTS
sales_by_category.write.mode("overwrite").saveAsTable("sales__category")
top_customers.write.mode("overwrite").saveAsTable("topcustomers")
sales_by_device.write.mode("overwrite").saveAsTable("sales_device")
delivery_analysis.write.mode("overwrite").saveAsTable("deliveryanalysis")
profit_by_category.write.mode("overwrite").saveAsTable("profit_category")
total_revenue.write.mode("overwrite").saveAsTable("totalrevenue")
```
